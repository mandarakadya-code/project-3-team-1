```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Creating a file
Co2=pd.read_csv('annual-co-emissions-by-region.csv')
Co2
```

```python
Co2.info()
```

```python
Co2.drop('Code',axis=1,inplace=True)
```

```python
Co2.rename(columns={'annual_co2_emissions':'A_Co2_emissions'},inplace=True)
```

```python
Co2.head()
```

```python
Co2.Year.unique()
```

```python
Co2.Entity.unique()
```

```python
# Change Entity to Country
Co2.rename(columns={'Entity':'Country'},inplace=True)
```

```python
Co2.info()
```

```python
# Creating a condition to separate each year
condition=Co2['A_Co2_emissions']>=0
df=Co2[condition]
```

```python
p1=df['Year']<=1959
```

```python
Period_1=df[p1]
```

```python
Period_1.head()
```

```python
#Creating Period of 10 years
P1A=df['Year']<=1970
P1B=df['Year']>=1960
```

```python
Period_2A=df[P1A&P1B]
Period_2A.head()
```

```python
# Combine Year and Co2 Emissions
Period_2A.groupby('Year')['A_Co2_emissions'].sum().sort_values(ascending=False)
```

```python
Period_2A.Year.unique()
```

```python
Period_2A.groupby('Year')['A_Co2_emissions'].sum().sort_values(ascending=False)
```

```python
# Creating a bar graph and pie chart from 1960-1970( Just trying to see wnat is the data of 10 years compare to 20 years)
temp=Period_2A[['Year','A_Co2_emissions']].groupby('Year',as_index=False).mean()
temp= temp.sort_values(['A_Co2_emissions'],ascending=False)

fig, (ax1,ax2)= plt.subplots(1,2)
fig.suptitle('1960-1970 Co2 Emisssions', fontweight='bold',fontsize=25)
temp.head(11).plot.bar(x='Year',color='blue',figsize=(18,15),fontsize=20,ax=ax1)
ax1.set_ylabel('A_Co2_emissions',fontweight='bold',fontsize=20)
ax1.set_xlabel('Year',fontweight='bold',fontsize=15)
temp.head(11).plot(kind='pie',x='Year',y='A_Co2_emissions',figsize=(30,20),startangle=90,shadow= {'ox': -0.09, 'edgecolor': 'none', 'shade': 0.11},labels=temp['Year'],legend=False,fontsize=20,autopct="%1.1f%%",ax=ax2)
ax2.set_ylabel('')
ax2.set_xlabel('')
fig.subplots_adjust(hspace=0.0)
```
